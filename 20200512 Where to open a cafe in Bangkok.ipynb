### Map data into a dataframe

```python
items = results['response']['groups'][0]['items']
items[0]
df1 = json_normalize(items) # flatten JSON
df1.head()
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# filter columns
filtered_columns = ['venue.name', 'venue.categories'] + [col for col in df1.columns if col.startswith('venue.location.')] + ['venue.id']
df1_filtered = df1.loc[:, filtered_columns]

# filter the category for each row
df1_filtered['venue.categories'] = df1_filtered.apply(get_category_type, axis=1)

# clean columns
df1_filtered.columns = [col.split('.')[-1] for col in df1_filtered.columns]

df1_filtered['name'] = df1_filtered['name'].str.replace(r"\(.*?\)","")
df1_filtered['name']
df1_filtered.head()

```

```python
cafe_df = df1_filtered[['name','categories','address','crossStreet','lat','lng','postalCode', 'city', 'id' ]]
#Combine address
cafe_df['location'] = cafe_df[['address','crossStreet']].astype(str).apply(lambda x: '/'.join(x), axis=1)
cafe_df = cafe_df.drop(cafe_df[['address','crossStreet']], axis = 1)

cafe_df = cafe_df[['name','categories','location','postalCode', 'city','lat','lng', 'id' ]]
cafe_df
```

### Clustering By Coordinates

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
kclusters = 12
cafe_ll = cafe_df[['lat','lng']]

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(cafe_ll)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:kclusters] 

# add clustering labels
try:
    cafe_df = cafe_df.drop(columns='Cluster')
except KeyError: 
    None
    
cafe_df.insert(0, 'Cluster', kmeans.labels_,)
cafe_df
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(cafe_df['lat'], cafe_df['lng'], cafe_df['name'], cafe_df['Cluster']):
    label = folium.Popup(str(poi) + ', Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

```python
cluster_number = []
number_of_cafe = []

for n in range(0,cafe_df['Cluster'].nunique()):
    cluster = cafe_df.loc[cafe_df['Cluster'] == n, cafe_df.columns[[1]]]
    cluster_size = len(cluster)
    cluster_number.append(n)
    number_of_cafe.append(cluster_size)
    
data = {'cluster_number':cluster_number,'number_of_cafe':number_of_cafe}
cluster_df = pd.DataFrame(data)
cluster_df
```

```python
##Change this to retrive cafes information in each cluster
cluster_n = 1

cluster1 = cafe_df.loc[cafe_df['Cluster'] == cluster_n]
cluster1 = cluster1.reset_index(drop = True)
cluster1
```

### Group by City

```python
city_grouped = cafe_df.groupby('city').nunique()
city_grouped = city_grouped.drop(columns='Cluster')
city_couunt = len(city_grouped)
print('There are {} cities'.format(city_couunt))
city_grouped = city_grouped.sort_values('location', ascending=False)
city_grouped['city_number'] = range(city_couunt)
```

```python
cafe_df_merged = cafe_df.join(city_grouped['city_number'], on ='city')
cafe_df_merged = cafe_df_merged.dropna()
cafe_df_merged['city_number'] = cafe_df_merged.city_number.astype(int)
cafe_df_merged = cafe_df_merged.drop('Cluster', axis = 1)
cafe_df_merged
```

```python
# create map
map_city = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x2 = np.arange(city_couunt)
ys2 = [i + x2 + (i*x2)**2 for i in range(city_couunt)]
colors_array2 = cm.rainbow(np.linspace(0, 1, len(ys2)))
rainbow2 = [colors.rgb2hex(i) for i in colors_array2]

# add markers to the map
markers_colors = []
for lat, lon, poi, city in zip(cafe_df_merged['lat'], cafe_df_merged['lng'], cafe_df_merged['name'], cafe_df_merged['city_number']):
    label = folium.Popup(str(poi) + ', in city group  '+str(city) , parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow2[city-1],
        fill=True,
        fill_color=rainbow2[city-1],
        fill_opacity=0.7).add_to(map_city)
       
map_city
```

```python
# a dataframe containing all information
city_directory = pd.DataFrame(cafe_df_merged, columns=['name','city', 'city_number'])

# make a new dataframe with just cities and ids
# drop duplicate rows
city_info = city_directory.loc[:, ['city','city_number']].drop_duplicates()

# get the cafe counts (number of cafes)
cafe_count = city_directory.groupby('city').count().iloc[:,0]

# add the cafe counts to the dataframe
city_info['cafe_count'] = cafe_count[city_info['city']].to_numpy()

# reset the index
city_info = city_info.sort_values('city_number', ascending=True)
city_info = city_info.reset_index(drop=True)

city_info
```
